```python
# Create a function to remove outlier   
def find_outliers(data, whisker_coeff = 1):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)                
    IQR = Q3 - Q1                                   
    lower = Q1 - whisker_coeff * IQR
    upper = Q3 + whisker_coeff * IQR
    return lower, upper

class RemoveOuliers():
    def __init__(self, whisker_coeff = 1):
        self.whisker = whisker_coeff
        self.lower = None
        self.upper = None
    def fit(self, X: pd.Series):
        self.lower, self.upper = find_outliers(X, self.whisker)
    def transform(self, X):
        return X.clip(self.lower, self.upper)
```

```python
for col in previous_application.columns:
    if previous_application[col].dtype != 'object':
        remove = RemoveOuliers()
        remove.fit(previous_application[col])
        previous_application[col] = remove.transform(previous_application[col])
```
