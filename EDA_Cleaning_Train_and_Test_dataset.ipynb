```python
import numpy as np
# Extract and preprocess age information
age_data = train[['TARGET', 'DAYS_BIRTH']].copy()
age_data['YEARS_BIRTH'] = age_data['DAYS_BIRTH'] / 365 * -1

# Bin the age data into 10 equal bins between 20 and 70 years
age_data['YEARS_BINNED'] = pd.cut(age_data['YEARS_BIRTH'], bins=np.linspace(20, 70, num=11))

# Group by age bins and calculate the mean of each group
age_groups = age_data.groupby('YEARS_BINNED', observed=False).mean()

# Plot the results
plt.figure(figsize=(10, 6))
plt.bar(age_groups.index.astype(str), 100 * age_groups['TARGET'], color='skyblue', edgecolor='black')
plt.xticks(rotation=75)
plt.xlabel('Age Group (years)')
plt.ylabel('Failure to Repay (%)')
plt.title('Failure to Repay by Age Group')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

From this chart, we can see that younger people tend to default on loan

# Cleaning

## Remove outliers

```python
# Create a function to remove outlier
def find_outliers(data, whisker_coeff = 1.5):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)                
    IQR = Q3 - Q1                                  
    lower = Q1 - whisker_coeff * IQR
    upper = Q3 + whisker_coeff * IQR
    return lower, upper


class RemoveOuliers():
    def __init__(self, whisker_coeff = 1.5):
        self.whisker = whisker_coeff
        self.lower = None
        self.upper = None
    def fit(self, X: pd.Series):
        self.lower, self.upper = find_outliers(X, self.whisker)
    def transform(self, X):
        return X.clip(self.lower, self.upper)
```

```python
# remove outliers in each columns in train dataset
for col in train.columns:
    if train[col].dtype != 'object':
        remove = RemoveOuliers()
        remove.fit(train[col])
        train[col] = remove.transform(train[col])
```

```python
# remove outliers in each columns in test dataset
for col in test.columns:
    if test[col].dtype not in ['object', 'category']:
        remove = RemoveOuliers()
        remove.fit(test[col])
        test[col] = remove.transform(test[col])
```

## Fill missing values

```python
from sklearn.impute import SimpleImputer
```

```python
# Fill missing values in both train and test dataset
for col in train.columns:
    if col != 'TARGET':
        if train[col].dtype in ['object', 'category']:
            imputer = SimpleImputer(strategy= 'most_frequent')
            imputer.fit(train[[col]])
            train[col] = imputer.transform(train[[col]]).ravel()
            test[col] = imputer.transform(test[[col]]).ravel()
        else:
            imputer = SimpleImputer(strategy= 'mean')
            imputer.fit(train[[col]])
            train[col] = imputer.transform(train[[col]]).ravel()
            test[col] = imputer.transform(test[[col]]).ravel()
        
```

## Encoding categorical variables

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot = OneHotEncoder(sparse_output=False, handle_unknown='ignore')  # Set sparse=False to get a dense array

# For the training dataset
categorical_columns = [col for col in train.columns if train[col].dtype in ['object', 'category']]  # Identify categorical columns
train_encoded = onehot.fit_transform(train[categorical_columns])  # Fit and transform only the categorical columns

# Replace the categorical columns with the one-hot encoded version
train = pd.concat([
    train.drop(columns=categorical_columns).reset_index(drop=True),  # Drop original categorical columns
    pd.DataFrame(train_encoded, columns=onehot.get_feature_names_out(categorical_columns))  # Add encoded columns
], axis=1)

# For the test dataset
test_encoded = onehot.transform(test[categorical_columns])  # Transform the test dataset with the same encoder

# Replace the categorical columns in the test dataset
test = pd.concat([
    test.drop(columns=categorical_columns).reset_index(drop=True),  # Drop original categorical columns
    pd.DataFrame(test_encoded, columns=onehot.get_feature_names_out(categorical_columns))  # Add encoded columns
], axis=1)
```
